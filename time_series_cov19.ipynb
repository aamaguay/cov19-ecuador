```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
import statsmodels.api as sm
import math
from statsmodels.tsa.arima_model import ARIMA
```

```python
url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
df = pd.read_csv (url, error_bad_lines = False)
```

```python
df_ec= df[df['Country/Region']=='Ecuador'].reset_index().iloc[:,5:].T
df_ec.rename({0: 'EC'}, axis=1, inplace=True)
df_ec.loc["3/14/20"]=28
```

```python
plt.figure(figsize = (25,15))
plt.plot(df_ec)
plt.xticks(rotation=45,size=16,color='black')
plt.title('Confirmed Cases in EC',size=25)
plt.xticks(size=16,color='black')
plt.yticks(size=17,color='black')
plt.xlabel("Days",size=17)
plt.ylabel("Confirmed Cases",size=17)
plt.savefig('EC_COV19', bbox_inches='tight', dpi=100) 
plt.show()
```
